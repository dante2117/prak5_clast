Второй по популярности являеется PlayStation 2 компании Sony
Третье место делят PlayStation Portable и PlayStation т той же Sony

```python

graph = sns.barplot(x=df["Platform"], y=df["Global_Sales"]) #объединение значений
graph.figure.set_figwidth(15)
graph.figure.set_figheight(10) 


```

Наибольшее количество мировых продаж видеоигр приходится на NES (Nintendo Entertainment System), второй идёт Wii (Nintendo Wii), третье место сложно точно определить, примерно равыне показатели имеют PS3, X360, PS4 Platform и GB

```python

graph = sns.boxplot(x="Genre", y="Global_Sales",data = df,palette='rainbow') #создание диаграммы размаха
graph.figure.set_figwidth(15)
graph.figure.set_figheight(10) 
```

По графику можно наблюдать, что самыми продаваемыми играми по жанру в среднем являются платформеры (Platform) и шутеры (Shooter), вторым идут экшены (Action), третьими идут спортивные (Sports). Самым непопулярным жанром являются пазлы (Puzzle).


```python
MyHeatMap = sns.heatmap(df.corr(numeric_only=True), vmin=-1, vmax=1, cmap= 'rainbow')#тепловая карта с вводом данныхи коэффицентом корреляции радужного цвета 
MyHeatMap.figure.set_figwidth(10)
MyHeatMap.figure.set_figheight(5)
```

По цветовой карте можно наблюдать, что наиболее велико отношение между NA_Sales (продаж в Северной Америке) и EU_Sales (продаж в Европе), наименьшее отрицательное соотношение имеет рейтинг к JP_Sales(продажи в Японии)

```python
sns.pairplot(df, hue = 'Platform') #построение попарных взаимосвязей в наборе данных с выделением платформ
#ИЛИ
graph = sns.PairGrid(df) #сетка подзаголовко
graph = graph.map_upper(sns.scatterplot) # тип графика для недиагонального (верхняя часть)
graph = graph.map_lower(sns.kdeplot) # тип графика для недиагонального (нижняя часть)
graph = graph.map_diag(sns.kdeplot, lw = 2) #  тип графика для диагонали
```

По данным графикам можем рассудить, что велик разброс рейтинга с течением лет (где наиболее преобладает PS и X360)
Чуть меньшую интенсивность имеет соотношение продаж в Японии с течением лет. Заметно, что прожали в разных регионах никак не связаны друг с другом

```python

df_mini = df[["NA_Sales","EU_Sales","JP_Sales"]][::50]
df_mini #Группировка по критерию продаж в регионах
```

```python


from sklearn.cluster import KMeans #модуль для кластеризации

model = KMeans(n_clusters=3) # описание модели

model.fit(df_mini) #  моделирование

all_predictions = model.predict(df_mini) # предсказание на всем наборе данных

df_mini["KMeans"] = all_predictions

df_mini
```

```python
sns.countplot(data = df_mini, y = "KMeans") #вывод распределённых 3 кластеров
```

Данные гистограммы показывают, что значения, которые отнесены к 0-му кластеру преобладают

```python
sns.pairplot(df_mini,hue="KMeans") 
```

Судя по графикам, в отношении продаж Америки к европейским преобладают знаечни я нулевого кластера, но так же нулевой кластер на всех диаграмм нахожится в разбросе у нулевых значений

```python
#Переназначение набора данных без столбца предсказаний KMeans

df_mini = df[["NA_Sales","EU_Sales","JP_Sales"]][::1]
df_mini
```

```python
#DBSCAN - более продвинутая версия KMeans
from sklearn.cluster import DBSCAN

model = DBSCAN()

model.fit(df_mini) #метод изучения кластеров

df_mini["DBSCAN"] = model.labels_

df_mini
```

```python
sns.countplot(data = df_mini, y = "DBSCAN")
```

```python
sns.pairplot(df_mini[::1],hue="DBSCAN")
```

Судя по крафику, значения кластера 0 выстраивается вдоль отметки нуля и довольно редко выходит вперёд. В то время как значения -1 кластера имеют чуть более широкий разброс
